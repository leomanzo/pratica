# Logica de programação passo a passo para solução do desafio:


```python
# Passo 1 - importar base de dados da pasta vendas
import pandas as pd
import plotly.express as px #importando biblioteca plotly

tabela_vendas = pd.read_excel("/content/drive/MyDrive/Python Ds/Vendas/Vendas+-+Base+de+Dados (1).xlsx")
display(tabela_vendas)
```

```python
# Passo 2 - Calcular o produto mais vendido (em quantidade)

tb_quantidade_produto = tabela_vendas.groupby("Produto").sum() #soma os intens iguals em "produto"
tb_quantidade_produto = tb_quantidade_produto[["Quantidade", "Valor Unitário"]] # filtro com 2x[[ especificar colunas separado com , ]]
tb_quantidade_produto = tb_quantidade_produto.sort_values(by="Quantidade", ascending=False ) #ou True para o constrario
display(tb_quantidade_produto)
```

```python
# Passo 3 - Calcular o produto mais vendido (em faturamento)

#metodo para criar uma nova coluna ["Faturamento"]
tabela_vendas["Faturamento"] = tabela_vendas["Quantidade"] * tabela_vendas["Valor Unitário"]

#sum da tabela
tb_faturamento_produto = tabela_vendas.groupby("Produto").sum()

#filtro do Faturemanto
tb_faturamento_produto = tb_faturamento_produto[["Faturamento"]].sort_values(by="Faturamento", ascending=False)

#output
display(tb_faturamento_produto)
```

```python
# Passo 4 - Calcular a loja/estado que mais vendeu (em faturamento)

tb_faturamento_loja = tabela_vendas.groupby("Loja").sum()
tb_faturamento_loja = tb_faturamento_loja[["Faturamento"]].sort_values(by="Faturamento", ascending=False)

display(tb_faturamento_loja)
```

```python
# Passo 5 - Calcular o ticket medio por loja/estado.

tabela_vendas["Ticket Médio"] = tabela_vendas["Valor Unitário"] #criando uma tabela Ticket Medio = a Valor unitario
tb_ticket_medio = tabela_vendas.groupby("Loja").mean(numeric_only = True) #criando a media dos numeros .mean(numeric_only = True)
tb_ticket_medio = tb_ticket_medio[["Ticket Médio"]].sort_values(by="Ticket Médio", ascending= False)
display(tb_ticket_medio)
```

```python
# Passo 6 - Criar um gráfico/dashboard da loja/estado que mais vendeu (em faturamento)

#           .line para tipo linha e etc ver menu existente
grafico = px.bar(tb_faturamento_loja, y= "Faturamento", x= tb_faturamento_loja.index) # montando uma tabela onde eixo X = Faturamento e Y= lojas.
grafico.show()
```

```python
# Passo 7 - email para o setor responsável

import smtplib
import email.message as em

def enviar_email():
    corpo_email = f"""
    <p>Prezados, tudo bem?</p>
    <p></p>
    <p>Segue em anexo as tabelas de faturamento por loja</p>

    <p><b>Faturamento por loja: </b></p>
    <p>{tb_faturamento_loja_formatado_html()}</p>

    <p><b>Quantidade vendida por produto: </b></p>
    <p>{tb_quantidade_produto_formatado_html()}</p>

    <p><b>Faturamento por produto: </b></p>
    <p>{tb_faturamento_produto_formatado_html()}</p>

    <p><b>Ticket Médio por loja: </b></p>
    <p>{tb_ticket_medio_formatado_html()}</p>


    <p>Obrigado</p>
    <p></p>
    <p>Att, Seu nome</p>
    """

    msg = em.Message()
    msg['Subject'] = "ASSUNTO" #ASSUNTO DO E-MAIL#
    msg['From'] = 'emailquevaienviar@gmail.com' #E-MAIL QUE VAI ENVIAR O E-MAIL#
    #
    msg['To'] = 'emailquevaireceber@gmail.com'#E-MAIL QUE VAI RECEBER
    password = '123456789' #SENHA DO E-MAIL QUE VAI ENVIAR --> usar a aba segurança do google --> verificar com 2 etapas -- > senhas de App -- > colar a senha aqui
    msg.add_header('Content-Type', 'text/html')
    msg.set_payload(corpo_email )

    s = smtplib.SMTP('smtp.gmail.com: 587')
    s.starttls()
    # Login Credentials for sending the mail
    s.login(msg['From'], password)
    s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    print('Email enviado')
    enviar_email()
```
